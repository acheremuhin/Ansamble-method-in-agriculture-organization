```python
import numpy as np
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
```

```python
Base = pd.read_excel(r"D://Наука/Статьи и работы начатые/Бэггинг, Бустинг и Случайный лес/Base.xlsx")
```

```python
Base = Base.dropna()
```

```python
Base_with_dummies = pd.get_dummies(Base, prefix='ОПФ_', columns=['ОПФ'])
```

```python
y = Base_with_dummies["Результат"]
X = Base_with_dummies.drop("Результат", axis = 1)
```

```python
Boost = AdaBoostClassifier()
```

```python
X = X.apply(pd.to_numeric, axis=0)
X.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
Boost.fit(X_train,y_train)
predict = Boost.predict(X_test).reshape(-1, 1)
accuracy_score(predict,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
AB = AdaBoostClassifier()
param_grid = {'n_estimators': range(10,100,10),'learning_rate': range(2,9,1),'algorithm': ["SAMME", "SAMME.R"]}
AB_GS = GridSearchCV(AB, param_grid, cv=5)
AB_GS.fit(X_train, y_train)
AB_GS.cv_results
```
